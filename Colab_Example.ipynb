```python
# prompt: read CountyHealthData into a dataframe

import pandas as pd

# Assuming the file is in your Google Drive, adjust the path accordingly
df = pd.read_csv('/content/CountyHealthData_2014-2015.csv')
# Replace 'CountyHealthData.csv' with the actual filename and path if necessary
df.head()
```

```python
# prompt: show me all the columns from the dataframe

df.columns
```

```python
# prompt: find the mean for each numerical column grouped by state and save into a new dataframe

import pandas as pd
grouped_means = df.groupby('State')['Deaths'].mean()
new_df = pd.DataFrame(grouped_means).reset_index()
new_df
```

```python
# prompt: List all columns that are floats

float_cols = df.select_dtypes(include=['float64']).columns
float_cols = list(float_cols)
```

```python
# prompt: for each column in floats_cols, find the average by state and assign it to a new dataframe

import pandas as pd
# Create an empty dictionary to store the results
results = {}

# Loop through each column in floats_cols
for col in float_cols:
  # Group by state and calculate the mean for the current column
  grouped_means = df.groupby('State')[col].mean()

  # Convert to a DataFrame and reset the index
  new_df = pd.DataFrame(grouped_means).reset_index()

  # Store the result in the dictionary
  results[col] = new_df
```

```python
new_df.sample(5)
```

```python
# after putting the prompt in the chat: 'group a dataframe by state and find average of each column'
float_cols.append('State')
df_filter = df[float_cols].copy()
```

```python
df_state_means = df_filter.groupby('State').agg('mean')
```

```python
# prompt: save the dataframe as a csv

df_state_means.to_csv('state_means.csv')
```
